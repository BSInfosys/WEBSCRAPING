```python
final= pd.DataFrame()
for j in range (1,130):
  url='https://www.ambitionbox.com/list-of-companies?campaign=desktop_nav&page={}'.format(j)
  headers = {'User-Agent': 'Mozilla/5.0(Windows NT 6.3; Win64; x64)AppleWebkit/537.36(KHTML,like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
  webpage=requests.get(url, headers=headers).text
  soup=BeautifulSoup(webpage, 'lxml')
  company_card=soup.find_all('div',class_='companyCardWrapper')
  name=[]
  rating=[]
  reviews=[]
  allinfo=[]
  highltrated=[]
  lowrated=[]
  ratedvalue=[]
  sallerys=[]
  interview=[]
  job=[]
  benifits=[]
  photo=[]
  for i in company_card:
    name.append(i.find('h2').text.strip())
    rating.append(i.find('span',class_='companyCardWrapper__companyRatingValue').text.strip())
    reviews.append(i.find('span',class_= 'companyCardWrapper__ActionCount').text.strip())
    allinfo.append(i.find('span',class_= 'companyCardWrapper__interLinking').text.strip())
    ratedvalue.append(i.find('span',class_= 'companyCardWrapper__ratingValues'))
    lowrated.append(i.find('span',class_= 'companyCardWrapper__ratingValues'))
    sallerys.append(i.find_all('span',class_= 'companyCardWrapper__ActionCount')[0].text.strip())
    interview.append(i.find_all('span',class_= 'companyCardWrapper__ActionCount')[1].text.strip())
    job.append(i.find_all('span',class_= 'companyCardWrapper__ActionCount')[2].text.strip())
    photo.append(i.find_all('span',class_= 'companyCardWrapper__ActionCount')[3].text.strip())
    benifits.append(i.find_all('span',class_= 'companyCardWrapper__ActionCount')[4].text.strip())
  d={'name':name,'rating':rating,'reviews':reviews,'allinfo':allinfo,'ratedvalue':ratedvalue,'sallerys':sallerys,'interview':interview,'job':job,'benifits':benifits,'photo':photo}
  df=pd.DataFrame(d)
  final=final.append(df,ignore_index=True)
```
