```python
import pandas as pd
import seaborn as sns
tips=sns.load_dataset('tips')
titanic=pd.read_csv('/content/titanic_data.csv/tested.csv')
flights=sns.load_dataset('flights')
iris=sns.load_dataset('iris')
```

1.Scatterplot (Numerical-Numerical)

```python
sns.scatterplot(tips,x='total_bill',y='tip',hue=df['Sex'],style=tips['smoker'])
```

2. Bar Plot (Numercal- Categorical)

```python
titanic.head()
```

```python
sns.barplot(titanic,x='Pclass',y='Fare',hue='Sex')
```

3. Box Plot (Numerical-Categorical)

```python
sns.boxplot(titanic,x='Sex',y='Age',hue='Survived')
```

Dist Plot (Numerical-Categorical)

```python
sns.distplot(titanic[titanic['Survived']==0]['Age'],hist=False)
sns.distplot(titanic[titanic['Survived']==1]['Age'],hist=False)
```

5. HeatMap (Categorical-Categorical)

```python
sns.heatmap(pd.crosstab(titanic['Pclass'],titanic['Survived']))
```

```python
(titanic.groupby('Pclass').mean()['Survived']*100).plot(kind='bar')
```

Cluster Map (Categorical-Categorical)

```python
sns.clustermap(pd.crosstab(titanic['SibSp'],titanic['Survived']))
```
